```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("P6-SuperStoreUS-2015.csv")
df
```

```python
df['Product Base Margin'].fillna(df['Product Base Margin'].mean(),inplace=True)
df.isnull().sum()
```

```python
df['Order Priority'].value_counts()
```

```python
df['Order Priority'].unique()
```

```python
df['Order Priority']=df['Order Priority'].replace("Critical","Critical")
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Order Priority',data=df)
plt.title('Count of order priority')
plt.show()
```

```python
df['Ship Mode'].value_counts()
```

```python
df['Ship Mode'].value_counts().index
```

```python
x = df['Ship Mode'].value_counts().index
y = df['Ship Mode'].value_counts().values
```

```python

plt.figure(figsize=(5,4))
plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
plt.legend(loc=2)
plt.show()
```

```python
# shipping mode analysis
plt.figure(figsize=(6,4))
sns.countplot(x="Ship Mode",data=df,hue="Product Category")
plt.show()
```

```python
# customer segment analysis
plt.figure(figsize=(6,4))
sns.countplot(x="Customer Segment",data=df)
plt.show()
```

```python
# maximum order quantity
plt.figure(figsize=(6,4))
category_sales=df.groupby('Product Category').agg({'Quantity ordered new':'sum','Sales':'sum'})
category_sales=pd.DataFrame(category_sales)
category_sales
sns.barplot(x=category_sales.index,y=category_sales['Quantity ordered new'],color='magenta')
plt.ylabel('Quantity Ordered')
plt.show()

# maximum sell category wise
sns.barplot(x=category_sales.index,y=category_sales['Sales'])
plt.show()
```

```python
# Supplies category wise
plt.figure(figsize=(8,6))
# sns.countplot(x="Product Category",data=df,hue="Product Sub-Category")
sns.countplot(x="Product Category",data=df[df["Product Category"]=="Office Supplies"],hue="Product Sub-Category")

plt.show()
```

```python
# Profit category wise
plt.figure(figsize=(6,4))
sns.barplot(x="Product Category",y="Profit",data=df,estimator='sum')
plt.show()
```

```python
# Region Wise sales Data
region_sales=df.groupby('Region').Sales.sum()
region_sales.plot(kind='bar',color='violet')
plt.ylabel('Total Sales')
plt.show()

```

```python
#  Sales State wise 
sales=df.groupby('State or Province').Sales.sum().nlargest(10)
sales.plot(kind='bar',color='magenta')
plt.ylabel('Total sales')
plt.show()
```

```python
# Minimum profit state wise
sales=df.groupby('State or Province').Profit.sum().nsmallest(10)
sales.plot(kind='barh',color='g')
plt.xlabel('Total Profit')
plt.show()
```

```python
# Maximum profit state wise
sales=df.groupby('State or Province').Profit.sum().nlargest(10)
sales.plot(kind='bar',color='g')
plt.ylabel('Total Profit')
plt.show()
```

```python
# Total sales cities wise
sales=df.groupby('City').Sales.sum().nlargest(10)
sales.plot(kind='bar')
plt.ylabel('Total sales')
plt.show()
```

```python
# Total profit cities wise
sales=df.groupby('City').Profit.sum().nlargest(10)
sales.plot(kind='bar',color='violet')
plt.ylabel('Total Profit')
plt.show()
```

```python
# Minimum profit cities wise
sales=df.groupby('City').Profit.sum().nsmallest(10)
sales.plot(kind='barh',color='violet')
plt.xlabel('Total Profit')
plt.show()
```

```python
#Top 10 customer name wise sales
customer=df.groupby('Customer Name').Sales.sum().nlargest(10)
customer.plot(kind='barh',color='g')
plt.show()
```
